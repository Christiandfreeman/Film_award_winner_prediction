```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.metrics import confusion_matrix
neighbors = KNeighborsClassifier(n_neighbors=50)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = .20, random_state=101)
```

```python
neighbors.fit(x_train,y_train)
x_predicted = neighbors.predict(x_test)
```

```python

print(accuracy_score(y_test,x_predicted))
```

```python
cross_score = cross_val_score(neighbors,x_test,y_test,cv=10 )
print(cross_score)
print('cross Val Socre: ',cross_score.mean())
```

```python
confusion_matrix(y_test,x_predicted)
```

```python
from sklearn import tree
decision_tree = tree.DecisionTreeClassifier()
```

```python
decision_tree.fit(x_train, y_train)
tree_predict = decision_tree.predict(x_test)

```

```python
tree_cross = cross_val_score(decision_tree,x_test, y_test, cv=10)
print(tree_cross)
print('Decision Tree Cross Validation Score: ',tree_cross.mean())
```

```python
print('Tree Confusion Matrix: ')
print(confusion_matrix(y_test, tree_predict))
```

```python
print('Accuaracy Score: ',accuracy_score(y_test,tree_predict))
```

```python
from sklearn.linear_model import LogisticRegression
```
